```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Load image and convert to grey.
img = cv.imread('/content/Aidar - Greencard 2020.jpg', -1)
```

```python
plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Resize
img_resized = cv.resize(img,(400,500))
plt.imshow(img_resized, 'gray')
```

```python
#Rotate
(h, w) = img.shape[:2]
# calculate the center of the image
center = (w / 2, h / 2)
 
angle90 = 90
angle180 = 180
angle270 = 270
 
scale = 1.0

M = cv.getRotationMatrix2D(center, angle270, scale)
rotated270 = cv.warpAffine(img, M, (w, h))

plt.imshow(rotated270, 'gray')
```

```python
#saving 
cv.imwrite('/content/Aidar - Greencard 2020 edited.jpg',rotated270)
```
